```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Node:

    def __init__(self, data, label='', _children=(), _op=''):
        self.data = float(data)
        self._grad = 0.0
        self._backprop = lambda: None
        self.label = label
        self._children = set(_children)
        self._op = _op
        
    def __repr__(self):
        return f"Node(data={self.data},grad={self._grad},label={self.label},op={self._op})"

    def __add__(self, other):
        other = other if isinstance(other, Node) else Node(other, '+i')
        
        out = Node(
            data=self.data + other.data,
            label='+o',
            _children=(self, other),
            _op='+'
        )
        
        def _backprop():
            self._grad += out._grad * 1
            other._grad += out._grad * 1
        
        out._backprop = _backprop
        return out

    def __radd__(self, other):
        return self + other

    def __neg__(self):
        return self * (-1)

    def __sub__(self, other):
        return self + (-other)

    def __rsub__(self, other):
        return self - other

    def __mul__(self, other):
        other = other if isinstance(other, Node) else Node(other, '*i')
        
        out = Node(
            data=self.data * other.data,
            label='*o',
            _children=(self, other),
            _op='*'
        )
        
        def _backprop():
            self._grad += out._grad * other.data # pattern is child._grad += parent._grad * child.grad()
            other._grad += out._grad * self.data
            
        out._backprop = _backprop
        return out

    def __rmul__(self, other):
        return self * other

    def __truediv__(self, other):
        out = self * other**(-1)
        return out
    
    def __pow__(self, other):
        assert isinstance(other, (int,float)), "node can only exponentiated by int/float"
        out = Node(self.data**other,'pow',(self, ), 'pow')

        def _backprop():
            self._grad += out._grad * (other * self.data ** (other-1))

        out._backprop = _backprop
        return out
    
    def tanh(self):
        data = (math.exp(2 * self.data) - 1) / (math.exp(2 * self.data) + 1)
        out = Node(data,'tanh',(self, ),'tanh')
        
        def _backprop():
            self._grad += out._grad * (1 - data**2)
            
        out._backprop = _backprop
        return out

    def exp(self):
        out = Node(math.exp(self.data),'exp',(self, ),'exp')

        def _backprop():
            self._grad += out._grad * out.data

        out._backprop = _backprop
        return out

    def backprop(self):
        sorted_topo = []
        visited = set()
        
        def topological_sort(node, sorted_topo, visited):
            if node not in visited:
                visited.add(node)
                for child in node._children:
                    topological_sort(child, sorted_topo, visited)
                sorted_topo.append(node)

        topological_sort(self, sorted_topo, visited)
        self._grad = 1.0
        for node in reversed(sorted_topo):
            node._backprop()
```

```python
from graphviz import Digraph

def trace(root):
    nodes, edges = set(), set()
    def build(node):
        if node not in nodes:
            nodes.add(node)
            for child in node._children:
                edges.add((child, node))
                build(child)
    build(root)
    return nodes, edges

def draw_dot(root):
    dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'})

    nodes, edges = trace(root)
    for node in nodes:
        uid = str(id(node))
        dot.node(name=uid, label=f"{node.label}|d={node.data:.2f}, g={node._grad:.2f}", shape='record')
        if node._op:
            dot.node(name=uid+node._op, label=node._op)
            dot.edge(uid+node._op, uid)

    for node1, node2 in edges:
        dot.edge(str(id(node1)), str(id(node2)) + node2._op)

    return dot
```

```python
def example_computation_graph():
    a = Node( 2, 'a')
    b = Node(-1, 'b')
    c = Node( 4, 'c')
    d = Node( 6, 'd')
    e = a + b; e.label='e' # 5
    f = c * d; f.label='f' # 24
    g = e * f; g.label='g' # 120; g = (a + b) * (c * d)

    g._grad = dg_dg = 1 # base case; _gradient of loss function wrt itself is always 1
    
    e._grad = dg_de = dg_dg * f.data # dg/de = dg/dg * dg/de and dg/de = f
    f._grad = dg_df = dg_dg * e.data # dg/df = dg/dg * dg/df and dg/df = e
    d._grad = dg_dd = dg_df * c.data # dg/dd = dg/df * df/dd and df/dd = c
    c._grad = dg_dc = dg_df * d.data # dg/dc = dg/df * df/dc and df/dc = d
    b._grad = dg_db = dg_de * 1      # dg/db = dg/de * de/db and de/db = 1
    a._grad = dg_da = dg_de * 1      # dg/da = dg/de * de/da and de/da = 1

    # pattern is child._grad = parent._grad * child.grad()
    
    return g
```

```python
graph = example_computation_graph()

draw_dot(graph)
```

```python
def nn_forward_exp():
    x1 = Node( 2, 'x1')
    x2 = Node( 0, 'x2')
    w1 = Node(-3, 'w1')
    w2 = Node( 1, 'w2')
    b  = Node(6.881, 'b')
    x1w1 = x1 * w1; x1w1.label = 'x1w1'
    x2w2 = x2 * w2; x2w2.label = 'x2w2'
    dot = x1w1 + x2w2; dot.label = 'dot'
    total = dot + b; total.label = 'total'
    out = ((2 * total).exp() - 1) / ((2 * total).exp() + 1)
    out.backprop()
    return out
```

```python
class Neuron:

    def __init__(self, n_in):
        self.w = [Node(random.uniform(-1,1),'w') for _ in range(n_in)]
        self.b = Node(random.uniform(-1,1),'b')

    def __call__(self, x):
        assert len(x) == len(self.w), "neuron input vector must be same length as weight vector"
        act = sum([w_ * x_ for w_, x_ in zip(self.w, x)]) + self.b 
        out = act.tanh()
        return out

    def parameters(self):
        return self.w + [self.b]

class Layer:
    
    def __init__(self, n_in, n_out):
        self.neurons = [Neuron(n_in) for _ in range(n_out)]

    def __call__(self, x):
        out = [n(x) for n in self.neurons]
        return out

    def parameters(self):
        _params = []
        for neuron in self.neurons:
            _params += neuron.parameters()
        return _params

class MLP:

    def __init__(self, n_in, n_outs):
        n = [n_in] + n_outs
        self.layers = [Layer(in_, out_) for in_, out_ in zip(n[:-1], n[1:])]

    def __call__(self, x):
        for layer in self.layers:
            x = layer(x)
        return x

    def parameters(self):
        _params = []
        for layer in self.layers:
            _params += layer.parameters()
        return _params

    def descent(self, a):
        for p in self.parameters():
            p.data -= a * p._grad

    def zero_grad(self):
        for p in self.parameters():
            p._grad = 0
```

```python
n = MLP(3, [4, 4, 1])

x = [
    [2, 3, -1],
    [3, 1, 0.5],
    [0.5, 1, 1],
    [1, 1, -1]
]

y = [1, -1, -1, 1]

epochs = 20
a = 0.01
```

```python
for e in range(epochs):
    ypred = [n(x_)[0] for x_ in x]
    loss = sum([(y_ - ypred_)**2 for y_, ypred_ in zip(y, ypred)])
    loss.backprop()
    n.descent(a)
    n.zero_grad()
    print(e, loss)
```
